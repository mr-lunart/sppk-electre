```python
# Mapping ranking prosesor ke bobot
processor_ranking = {
    "AMD Ryzen 9": 100,
    "Intel Core i9": 100,
    "AMD Ryzen 7": 80,
    "Intel Core i7": 60,
    "AMD Ryzen 5": 40,
    "Intel Core i5": 40,
    "Intel Core i3": 20
}

# Menambahkan kolom ranking ke dataset
df['Processor'] = df['Processor'].map(processor_ranking)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head(5)
```

```python
# Mapping ranking prosesor ke bobot
# Definisikan fungsi untuk mengkategorikan harga
def categorize_price(price):
    if price < 10000000:
        return 20
    elif 10000000 <= price < 20000000:
        return 40
    elif 20000000 <= price < 30000000:
        return 60
    elif 30000000 <= price < 40000000:
        return 80
    elif price >= 40000000:
        return 100

# Terapkan fungsi ke kolom 'harga'
df['Harga'] = df['Harga'].apply(categorize_price)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head()
```

```python
# Mapping ranking prosesor ke bobot
def categorize_penyimpanan(simpan):
    if simpan == "4TB":
        return 100
    elif simpan == "2TB":
        return 80
    elif simpan == "1TB":
        return 60
    elif simpan == "512GB":
        return 40
    elif simpan == "256GB":
        return 20

# Terapkan fungsi ke kolom 'harga'
df['Penyimpanan'] = df['Penyimpanan'].apply(categorize_penyimpanan)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head()
```

```python
# Mapping ranking prosesor ke bobot
def categorize_layar(layar):
    if layar == "13 Inch":
        return 20
    elif layar == "14 Inch":
        return 40
    elif layar == "15 Inch":
        return 60
    elif layar == "16 Inch":
        return 80
    elif layar == "17 Inch":
        return 100
    elif layar == "18 Inch":
        return 100

# Terapkan fungsi ke kolom 'harga'
df['Ukuran Layar'] = df['Ukuran Layar'].apply(categorize_layar)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head()
```

```python
# Mapping ranking prosesor ke bobot
def categorize_keyboard(keyboard):
    if keyboard == "Standart Keyboard":
        return 50
    elif keyboard == "Backlit Keyboard":
        return 100

# Terapkan fungsi ke kolom 'harga'
df['Keyboard'] = df['Keyboard'].apply(categorize_keyboard)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head()
```

```python
# Mapping ranking prosesor ke bobot
def categorize_luayara(luayare):
    if luayare == "Non Touchscreen":
        return 50
    elif luayare == "Touchscreen":
        return 100

# Terapkan fungsi ke kolom 'harga'
df['Tipe Layar'] = df['Tipe Layar'].apply(categorize_luayara)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head()
```

```python
# Mapping ranking prosesor ke bobot
def categorize_ram(ram):
    if ram == "64GB":
        return 100
    elif ram == "32GB":
        return 80
    elif ram == "16GB":
        return 60
    elif ram == "8GB":
        return 40
    elif ram == "4GB":
        return 20

# Terapkan fungsi ke kolom 'harga'
df['RAM'] = df['RAM'].apply(categorize_ram)

# Tampilkan 5 baris pertama dari dataset yang sudah diubah untuk verifikasi
df.head()
```
